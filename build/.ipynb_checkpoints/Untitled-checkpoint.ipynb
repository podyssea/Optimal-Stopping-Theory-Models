```python
import pandas as pd
import seaborn as sns
import numpy as np
from matplotlib import pyplot as plt
from numpy import genfromtxt
import statsmodels.api as sm
from pandas.plotting import autocorrelation_plot

data_2 = pd.read_csv("ec2_cpu_data_second.csv")

x_labels_timestamp = data_2.iloc[:, 0]
positions = data_2.iloc[:, 1]
labels = x_labels_timestamp

del data_2["timestamp"]

print(data_2.diff())
differences = data_2.diff()
data_2['differences'] = differences

print(data_2.corr())
print(data_2.head())

#This plots X[t]
ax = autocorrelation_plot(data_2.iloc[:,0])
ax.set_xlim([0, 4033])
ax.set_ylim([-0.3,1])
```

```python
#This plots the Y[t]
plt.plot(differences)
```

```python
from scipy import stats, integrate

kde = stats.gaussian_kde(data_2.iloc[:,0])

# Calculate the integral of the kde between 10 and 20:
xmin, xmax = 0, 4033
integral, err = integrate.quad(kde, xmin, xmax)

x = np.linspace(0,4033,4033)
x_integral = np.linspace(xmin, xmax, 4033)

plt.plot(x, kde(x), label="KDE")
plt.fill_between(x_integral, 0, kde(x_integral),
                 alpha=0.3, color='b', label="Area: {:.3f}".format(integral))
plt.legend()
```
