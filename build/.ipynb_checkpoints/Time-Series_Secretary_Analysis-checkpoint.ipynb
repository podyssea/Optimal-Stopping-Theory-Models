# Code for the Secretary Problem

```python
#This is the code for the Secretary Problem

#From the dataframe these are the load values and the dates columns stored in arrays to be processed
distribution = df.reset_index(drop=False).iloc[:,1].values
dates = df.reset_index(drop=False).iloc[:,0].values

#Empty lists to store times we stopped, loads at each stop and the minimum values of each run
time_list = []
load_list = []
minimums = []

run = 2

#distribution = loads #np.random.uniform(0.1, 1.0, 1000) #
#minutes = list(range(4031))

#Graph of Observations
#ax = plt.subplots(1, 1,figsize=(20,10))
#line, = plt.plot(dates, distribution, lw=2)
#plt.show()

#Start by observing the first 100 values of the distribution for both loads and dates, compute the minimum and append it to minimums list
candidate = distribution[:100]
candidate_time = dates[:100]
minimumL = min(candidate)
minimums.append(minimumL)

#Compute the sample size and the first 37 observations
samplesize = round(len(candidate) * math.exp(-1))
sample = candidate[ : samplesize]

#print("RUN 1\n------------------------------------------------------")
#print ("Sample size is " + str(samplesize))

#Compute the benchmark of the sample (minimum values of the sample)
benchmark = min(sample)
#print ("Minimum of the sample is " + str(benchmark))
#print ("Minimum of the 100 observations is " + str(minimumL))
best = 0
index = samplesize

#Find the best candidate for the model to stop in the first run by comparing all next values with the benchmark value
while index <= len(candidate)-1:
    if candidate[index] <= benchmark:
        best = index
        break
    index += 1

#Once we observe the first that is less than the benchmark value STOP there and offload. Store the value
if(candidate[best] <= benchmark):
    #print ("Best candidate found! We offload on " + str(candidate_time[best]) + "\nThe load when we offload is " + str(candidate[best]))
    time_list.append(candidate_time[best])
    load_list.append(candidate[index])
    #print("The difference between the Optimal and the Achieved load values is " + str(candidate[best] - minimumL))

#If we dont then go to the end of the 100 observations, Stop there and offload
else:
    #print ("We offload on " + str(candidate_time[-1]) + "\nThe load when we offload is " + str(candidate[-1]))
    time_list.append(candidate_time[-1])
    load_list.append(candidate[samplesize])
    #print("The difference between the Optimal and the Achieved load values is " + str(candidate[-1] - minimumL))


#Prepare the next observations to be processed
remaining_observations = distribution[index+1:]
remaining_times = dates[index+1:]
next_observations = remaining_observations[:100]
next_times = remaining_times[:100]


#Repeat the process for every RUN
while(len(remaining_observations) != 0):
    
    samplesize = round(len(next_observations) * math.exp(-1))
    
    minimumL = min(next_observations)
    
    #print("\n------------------------------------------------------")
    #print("RUN " + str(run) + "\n------------------------------------------------------")
    #print ("Sample size is " + str(samplesize))

    sample = next_observations[ : samplesize]
    
    if (len(sample) == 0):
        break

    benchmark = min(sample)
    
    #print ("Minimum of the sample is " + str(benchmark))
    #print ("Minimum of the 100 observations is " + str(minimumL))

    best = 0
    index = samplesize
    while index <= len(next_observations)-1:
        if next_observations[index] <= benchmark:
            best = index
            break
        index += 1
    
    #Stop if remaining observations is empty
    if len(remaining_observations) == 0:
        break
    elif(next_observations[best] <= benchmark):
        #print ("Best candidate found! We offload on " + str(next_times[best]) + "\nThe load when we offload is "  + str(next_observations[best]))
        time_list.append(next_times[best])
        load_list.append(next_observations[best])
        #print("The difference between the Optimal and the Achieved load values is " + str(next_observations[best] - minimumL))
    else:
        #print ("We offload on " + str(next_times[-1]) + "\nThe load when we offload is " + str(next_observations[-1]))
        time_list.append(next_times[-1])
        load_list.append(next_observations[-1])
        #print("The difference between the Optimal and the Achieved load values is " + str(next_observations[-1] - minimumL))

    remaining_observations = remaining_observations[index+1:]
    remaining_times = dates[index+1:]
    next_observations = remaining_observations[:100]
    next_times = remaining_times[:100]
    
        
    run += 1
    
    #Stop if next observations is empty
    if len(next_observations) == 0:
        break
    else:
        minimums.append(minimumL)

#Convert timedeltas (difference) to minutes
time_list.insert(0, dates[0])
steps = np.array([i-j for i, j in zip(time_list[:-1], time_list[1:])])
time_differences = [int(x) for x in steps]
time_differences = np.absolute(np.asarray(time_differences)/60000000000)

load_differences = [a - b for a, b in zip(load_list, minimums)]
```

# Analysis of the Results and Plots

```python
# data to plot
n_groups = len(minimums)

# create plot
fig, ax = plt.subplots(1, 1,figsize=(20,10)) # plt.subplots()
index = np.arange(n_groups)
bar_width = 0.4
opacity = 0.8

#Plot the minimum values of each observed sample
rects1 = plt.bar(index, minimums, bar_width,
alpha=opacity,
color='b',
label='Optimal')

#Plot the achieved values of each observed samle
rects2 = plt.bar(index + bar_width, load_list, bar_width,
alpha=opacity,
color='g',
label='Achieved')

#Label
plt.xlabel('Stops')
plt.ylabel('Load Values')
plt.title('Loads by Stop')
plt.xticks(index + bar_width, tuple(range(1,len(time_list)+1)))
plt.legend()

#plt.tight_layout()
plt.show()

#Display the dataframes
runs_data = {'Run':  list(range(1,len(minimums)+1)),
             'Optimal': minimums,
            'Load when Offloading': load_list,
            'Load Difference': load_differences,
            'Stop Time Between Runs in Mins': time_differences
            }

analysis_data = {'Data': ['Loads','Times'], 'Mean': [np.mean(load_list), np.mean(np.absolute(np.asarray(time_differences)/60000000000))],
                'Variance': [np.var(load_list), np.var(np.absolute(np.asarray(time_differences)/60000000000))]
                }

runs_frame = pd.DataFrame(runs_data, columns = ['Run','Optimal','Load when Offloading', 'Load Difference',
                                                'Stop Time Between Runs in Mins'])

runs_frame.index += 1

analysis_frame = pd.DataFrame(analysis_data, columns = [ 'Data','Mean', 'Variance'])                

analysis_frame.index += 1
                          
display(runs_frame)
display(analysis_frame)

```
