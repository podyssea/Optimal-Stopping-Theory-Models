# Code for the Random(p) Problem

```python
#This is the code for the Random Probability
#Set the counter and the probability before proceeding to the next block of code
#Define counter = N, probability = 0.05
def random_prob_model(counter, probability):
    global next_times
    #From the dataframe these are the load values and the dates columns stored in arrays to be processed
    distribution = df.reset_index(drop=False).iloc[:,1].values
    dates = df.reset_index(drop=False).iloc[:,0].values

    #Fix a probability
    #probability = 0.5
    #counter = 100

    #Empty lists to store times we stopped, loads at each stop and the minimum values of each run and times
    time_list = []
    load_list = []
    minimums = []
    minimums_times = []

    run = 1

    #Start by observing the first 100 values of the distribution for both loads and dates, compute the minimum and append it to minimums list
    candidate = distribution[:counter]
    candidate_time = dates[:counter]
    minimumL = min(candidate)

    #Compute the minimum of the next 100 observations
    best = 0
    index = 0

    #Find the best candidate for the model to stop in the first run by comparing the random generated value with the fixed probability x
    #If less STOP and offload, else Keep looking unitl the 100th observation
    while index <= len(candidate)-1:
        x = random.uniform(0, 1)
        if x < probability:
            best = index
            #print ("Best candidate found! We offload on " + str(candidate_time[best]) + "\nThe load when we offload is " + str(candidate[best]))
            time_list.append(candidate_time[best])
            load_list.append(candidate[best])
            #print("The difference between the Optimal and the Achieved load values is " + str(candidate[best] - minimumL))
            break
        elif index == counter-1:
            best = index
            #print ("Best candidate found! We offload on " + str(next_times[-1]) + "\nThe load when we offload is "  + str(next_observations[-1]))
            time_list.append(candidate_time[best])
            load_list.append(candidate[best])
            #print("The difference between the Optimal and the Achieved load values is " + str(next_observations[-1] - minimumL))
            break
        index += 1

    #Prepare the next observations to be processed
    remaining_observations = distribution[counter:]
    remaining_times = dates[counter:]
    next_observations = remaining_observations[:counter]
    next_times = remaining_times[:counter]
    minimums.append(minimumL)
    minimums_times.append(dates[np.where(distribution == minimumL)][0])
    #print("Appended value",minimumL, "in minimums, ", candidate_time[best], "in time_list and ", candidate[best], "in load_list in run", run)

    run = 2

    #Repeat the process for every RUN every 100 observations
    while(len(next_observations) != 0):

        minimumL = min(next_observations)
        
        #Set the next time observation set
        remaining_observations = remaining_observations[counter:]
        remaining_times = dates[counter:]
        next_observations = remaining_observations[:counter]
        next_times = remaining_times[:counter]
#         minimums.append(minimumL)
#         minimums_times.append(dates[np.where(distribution == minimumL)][0])
        
        #print("\n------------------------------------------------------")
        #print ("Minimum of the 100 observations is " + str(minimumL))

        best = 0
        index = 0

        while index <= len(next_observations)-1:

            x = random.uniform(0, 1)
            #print(x)
            if x < probability:
                best = index
                if len(remaining_observations) == 0:
                    break
                else:
                    #print ("Best candidate found! We offload on " + str(next_times[best]) + "\nThe load when we offload is "  + str(next_observations[best]))
                    time_list.append(next_times[best])
                    load_list.append(next_observations[best])
                    #print("The difference between the Optimal and the Achieved load values is " + str(next_observations[best] - minimumL))
                    minimums.append(minimumL)
                    minimums_times.append(dates[np.where(distribution == minimumL)][0])
                    break
            elif index == len(next_observations)-1:
                best = index
                #print ("Best candidate found! We offload on " + str(next_times[-1]) + "\nThe load when we offload is "  + str(next_observations[-1]))
                time_list.append(next_times[best])
                load_list.append(next_observations[best])
                #print("The difference between the Optimal and the Achieved load values is " + str(next_observations[-1] - minimumL))
                minimums.append(minimumL)
                minimums_times.append(dates[np.where(distribution == minimumL)][0])
                break

            index += 1
            
#         #Set the next time observation set
#         remaining_observations = remaining_observations[counter:]
#         remaining_times = dates[counter:]
#         next_observations = remaining_observations[:counter]
#         next_times = remaining_times[:counter]
#         minimums.append(minimumL)
#         minimums_times.append(dates[np.where(distribution == minimumL)][0])
        
        
        if len(next_observations) == 0:
            break  
        #print("Appended value",minimumL, "in minimums, ", next_times[best], "in time_list and ", next_observations[best], "in load_list in run", run, "with size",len(next_observations))

        
        run += 1
    
    #Convert timedeltas (difference) to minutes of achieved times
    #time_list.insert(0, dates[0])
    steps = np.array([i-j for i, j in zip(time_list[:-1], time_list[1:])])
    time_differences = [int(x) for x in steps]
    time_differences = np.absolute(np.asarray(time_differences)/1e-9)

    #Convert timedeltas (diff) to minutes of optimal times
    #minimums_times.insert(0, dates[0])
    minimums_steps = np.array([i-j for i, j in zip(minimums_times[:-1], minimums_times[1:])])
    minimums_time_differences = [int(x) for x in minimums_steps]
    minimums_time_differences = np.absolute(np.asarray(minimums_time_differences)/1e-9)

    load_differences = [a - b for a, b in zip(load_list, minimums)]
    times_diff = np.array(time_list) - np.array(minimums_times)

    return minimums, load_list, load_differences, minimums_times, time_list, times_diff
    #print("-------------------------------------------------------------")

#ALL COMBINATIONS OF CHUNKS AND PROBABILITIES
#RUN THESE BEFORE PROCEEDING TO ANALYSIS
rpb_20 = [random_prob_model(20, 0.05),
random_prob_model(20, 0.1),
random_prob_model(20, 0.2),
random_prob_model(20, 0.3),
random_prob_model(20, 0.5)]

rpb_50 = [random_prob_model(50, 0.05),
random_prob_model(50, 0.1),
random_prob_model(50, 0.2),
random_prob_model(50, 0.3),
random_prob_model(50, 0.5)]

rpb_80 = [random_prob_model(80, 0.05),
random_prob_model(80, 0.1),
random_prob_model(80, 0.2),
random_prob_model(80, 0.3),
random_prob_model(80, 0.5)]

rpb_100 = [random_prob_model(100, 0.05),
random_prob_model(100, 0.1),
random_prob_model(100, 0.2),
random_prob_model(100, 0.3),
random_prob_model(100, 0.5)]

rpb_150 = [random_prob_model(150, 0.05),
random_prob_model(150, 0.1),
random_prob_model(150, 0.2),
random_prob_model(150, 0.3),
random_prob_model(150, 0.5)]

rpb_200 = [random_prob_model(200, 0.05),
random_prob_model(200, 0.1),
random_prob_model(200, 0.2),
random_prob_model(200, 0.3),
random_prob_model(200, 0.5)]
```

# Code for the Secretary Problem

```python
#This is the code for the Secretary  Problem
#Set the counter and the probability before proceeding to the next block of code
#Define counter = N
def secretary_model(counter):
    #This is the code for the Secretary Problem

    #From the dataframe these are the load values and the dates columns stored in arrays to be processed
    distribution = df.reset_index(drop=False).iloc[:,1].values
    dates = df.reset_index(drop=False).iloc[:,0].values
    #print("-------------------------------------------------------------")
    #print("Results N =", counter)

    #Empty lists to store times we stopped, loads at each stop and the minimum values of each run
    time_list_secretary = []
    load_list_secretary = []
    minimums_secretary = []
    minimums_times_secretary = []

    run_secretary = 2

    #distribution = loads #np.random.uniform(0.1, 1.0, 1000) #
    #minutes = list(range(4031))

    #Graph of Observations
    #ax = plt.subplots(1, 1,figsize=(20,10))
    #line, = plt.plot(dates, distribution, lw=2)
    #plt.show()

    #Start by observing the first 100 values of the distribution for both loads and dates, compute the minimum and append it to minimums_secretary list
    candidate_secretary = distribution[:counter]
    candidate_time_secretary = dates[:counter]
    minimumL_secretary = min(candidate_secretary)
    minimums_secretary.append(minimumL_secretary)
    minimums_times_secretary.append(dates[np.where(distribution == minimumL_secretary)][0])


    #Compute the sample size and the first 37 observations
    samplesize_secretary = round(len(candidate_secretary) * math.exp(-1))
    sample_secretary = candidate_secretary[ : samplesize_secretary]

    #print("RUN 1\n------------------------------------------------------")
    #print ("Sample size is " + str(samplesize_secretary))

    #Compute the benchmark_secretary of the sample (minimum values of the sample)
    benchmark_secretary = min(sample_secretary)
    #print ("Minimum of the sample is " + str(benchmark_secretary))
    #print ("Minimum of the 100 observations is " + str(minimumL_secretary))
    best_secretary = 0
    index_secretary = samplesize_secretary

    #Find the best_secretary candidate_secretary for the model to stop in the first run by comparing all next values with the benchmark_secretary value
    while index_secretary <= len(candidate_secretary)-1:
        if candidate_secretary[index_secretary] <= benchmark_secretary:
            best_secretary = index_secretary
            break
        index_secretary += 1

    #Once we observe the first that is less than the benchmark_secretary value STOP there and offload. Store the value
    if(candidate_secretary[best_secretary] <= benchmark_secretary):
        #print ("best_secretary candidate_secretary found! We offload on " + str(candidate_time_secretary[best_secretary]) + "\nThe load when we offload is " + str(candidate_secretary[best_secretary]))
        time_list_secretary.append(candidate_time_secretary[best_secretary])
        load_list_secretary.append(candidate_secretary[index_secretary])
        #print("The difference between the Optimal and the Achieved load values is " + str(candidate_secretary[best_secretary] - minimumL_secretary))

    #If we dont then go to the end of the 100 observations, Stop there and offload
    else:
        #print ("We offload on " + str(candidate_time_secretary[-1]) + "\nThe load when we offload is " + str(candidate_secretary[-1]))
        time_list_secretary.append(candidate_time_secretary[-1])
        load_list_secretary.append(candidate_secretary[samplesize_secretary])
        #print("The difference between the Optimal and the Achieved load values is " + str(candidate_secretary[-1] - minimumL_secretary))


    #Prepare the next observations to be processed
    remaining_observations_secretary = distribution[index_secretary+1:]
    remaining_times_secretary = dates[index_secretary+1:]
    next_observations_secretary = remaining_observations_secretary[:counter]
    next_times_secretary = remaining_times_secretary[:counter]


    #Repeat the process for every RUN
    while(len(remaining_observations_secretary) != 0):

        samplesize_secretary = round(len(next_observations_secretary) * math.exp(-1))

        minimumL_secretary = min(next_observations_secretary)


        #print("\n------------------------------------------------------")
        #print("RUN " + str(run) + "\n------------------------------------------------------")
        #print ("Sample size is " + str(samplesize_secretary))

        sample_secretary = next_observations_secretary[ : samplesize_secretary]

        if (len(sample_secretary) == 0):
            break

        benchmark_secretary = min(sample_secretary)

        #print ("Minimum of the sample is " + str(benchmark_secretary))
        #print ("Minimum of the 100 observations is " + str(minimumL_secretary))

        best_secretary = 0
        index_secretary = samplesize_secretary
        while index_secretary <= len(next_observations_secretary)-1:
            if next_observations_secretary[index_secretary] <= benchmark_secretary:
                best_secretary = index_secretary
                break
            index_secretary += 1

        #Stop if remaining observations is empty
        if len(remaining_observations_secretary) == 0:
            break
        elif(next_observations_secretary[best_secretary] <= benchmark_secretary):
            #print ("best_secretary candidate_secretary found! We offload on " + str(next_times_secretary[best_secretary]) + "\nThe load when we offload is "  + str(next_observations_secretary[best_secretary]))
            time_list_secretary.append(next_times_secretary[best_secretary])
            load_list_secretary.append(next_observations_secretary[best_secretary])
            #print("The difference between the Optimal and the Achieved load values is " + str(next_observations_secretary[best_secretary] - minimumL_secretary))
        else:
            #print ("We offload on " + str(next_times_secretary[-1]) + "\nThe load when we offload is " + str(next_observations_secretary[-1]))
            time_list_secretary.append(next_times_secretary[-1])
            load_list_secretary.append(next_observations_secretary[-1])
            #print("The difference between the Optimal and the Achieved load values is " + str(next_observations_secretary[-1] - minimumL_secretary))

        remaining_observations_secretary = remaining_observations_secretary[index_secretary+1:]
        remaining_times_secretary = dates[index_secretary+1:]
        next_observations_secretary = remaining_observations_secretary[:counter]
        next_times_secretary = remaining_times_secretary[:counter]

        run_secretary += 1
        counter += counter

        minimums_secretary.append(minimumL_secretary)
        minimums_times_secretary.append(dates[np.where(distribution == minimumL_secretary)][0])

        #Stop if next observations is empty
        if len(next_observations_secretary) == 0:
            break
        #else:
        #    minimums_secretary.append(minimumL_secretary)

    #Convert timedeltas (difference) to minutes
    #time_list_secretary.insert(0, dates[0])
    steps_secretary = np.array([i-j for i, j in zip(time_list_secretary[:-1], time_list_secretary[1:])])
    time_differences_secretary = [int(x) for x in steps_secretary]
    time_differences_secretary = np.absolute(np.asarray(time_differences_secretary)/1e-9)

    #NEEDS REVIEWING
    #Convert timedeltas (diff) to minutes of optimal times
    #minimums_times_secretary.insert(0, dates[0])
    minimums_steps_secretary = np.array([i-j for i, j in zip(minimums_times_secretary[:-1], minimums_times_secretary[1:])])
    minimums_times_differences_secretary = [int(x) for x in list(minimums_steps_secretary)]
    minimums_times_differences_secretary = np.absolute(np.asarray(minimums_times_differences_secretary)/1e-9)
    
    load_differences_secretary = [a - b for a, b in zip(load_list_secretary, minimums_secretary)]
    times_diff_secretary = np.array(time_list_secretary) - np.array(minimums_times_secretary)

    return minimums_secretary, load_list_secretary, load_differences_secretary, minimums_times_secretary, time_list_secretary, times_diff_secretary
    #print("-------------------------------------------------------------")
```

# RANDOM AND SECRETARY MODELS Vs OPTIMAL Task 2

```python
#Plot the different models (random(P) for different probabilities and seecretary model) to compare with the optimal for each model

#Set the rpb_model(eg. rpb_200) and the secretary model(eg. secretary model(200))
def avg_loads_by_stop(rpb_model, secretary_model):
    fig, ax = plt.subplots(1, 1,figsize=(15,15)) # plt.subplots()
    bar_width = 0.4
    opacity = 0.8
    
    optimal_means = [np.mean(rpb_model[0][0]),np.mean(rpb_model[1][0]),np.mean(rpb_model[2][0]),np.mean(rpb_model[3][0]),
                              np.mean(rpb_model[4][0]), np.mean(secretary_model[0])]
    
    achieved_means = [np.mean(rpb_model[0][1]),np.mean(rpb_model[1][1]),np.mean(rpb_model[2][1]),np.mean(rpb_model[3][1]),
                             np.mean(rpb_model[4][1]),np.mean(secretary_model[1])]
    
    all_means = [np.mean(rpb_model[0][0]), np.mean(secretary_model[0]), np.mean(rpb_model[0][1]),np.mean(rpb_model[1][1]),np.mean(rpb_model[2][1]),np.mean(rpb_model[3][1]),
                             np.mean(rpb_model[4][1]),np.mean(secretary_model[1])]

    #Plot the minimum values of each observed sample
#     rects1 = plt.bar([0,1,2,3,4,5], optimal_means , bar_width,
#     alpha=opacity,
#     color='b',
#     label='Optimal')

    #Plot the achieved values of each observed samle
    rects2 = plt.bar(np.arange(8) + bar_width, all_means, bar_width,alpha=opacity,color = '#99ccff',label='Means')

    #Label
    x_ticks_labels = ['Random(P) Optimal', 'Secretary Optimal', 'Random(P = 0.05)','Random(P = 0.1)','Random(P = 0.2)','Random(P = 0.3)','Random(P = 0.5)', 'Secretary']
    plt.xlabel('Models')
    plt.ylabel('Load Values')
    plt.title('Avg Loads by Stop for each Model with N = 200')
    plt.xticks(np.arange(8) + bar_width, ('Random(P) Optimal', 'Secretary Optimal','Random(P = 0.05)','Random(P = 0.1)','Random(P = 0.2)','Random(P = 0.3)','Random(P = 0.5)', 'Secretary'), rotation = 45)
    for p in ax.patches:
            ax.annotate("%.2f" % p.get_height(), (p.get_x() + p.get_width() / 2., p.get_height()),
                 ha='center', va='center', rotation=0, xytext=(0, 20), textcoords='offset points')
    #plt.savefig('Averages for N = 200.png')
    plt.legend()

    #plt.tight_layout()
    plt.show()
    
    #Display the dataframe
    runs_data = {'Model':  ['Random(P = 0.05)','Random(P = 0.1)','Random(P = 0.2)','Random(P = 0.3)','Random(P = 0.5)', 'Secretary'],
                 'Optimal Means': optimal_means,
                 'Offloading Means': achieved_means,
                 'Mean Load Difference': np.array(achieved_means) - np.array(optimal_means)}

    runs_frame = pd.DataFrame(runs_data, columns = ['Model','Optimal Means',
                                                    'Offloading Means',
                                                    'Mean Load Difference'])
    runs_frame.index += 1
    
    
    
    fig = plt.figure(figsize=(20,20))
    ax1 = plt.subplot(111)
    ret = ax1.bar(runs_frame['Model'], runs_frame['Mean Load Difference'], color = '#99ccff')
    ret[np.where(runs_frame['Mean Load Difference'] == runs_frame['Mean Load Difference'].min())[0][0]].set_color('#404040')
    plt.xlabel('Models')
    plt.ylabel('Load Difference')
    plt.title('Load Mean Differences for N = 200')
    for p in ax1.patches:
            ax1.annotate("%.2f" % p.get_height(), (p.get_x() + p.get_width() / 2., p.get_height()),
                 ha='center', va='center', rotation=0, xytext=(0, 20), textcoords='offset points')
    
    
    #plt.savefig('Best for N = 200.png')
    plt.show
    
    dfi.export(runs_frame, 'runs_frame_200.png')
    
    display(runs_frame)

    
```
